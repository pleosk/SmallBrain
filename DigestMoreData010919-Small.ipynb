```python
import pandas as pd 
#dft = pd.read_csv("MoreData/dft.csv",sep='\t') 
dft = pd.read_csv("Commit/dfts.csv",sep='\t') 
# Preview the first 5 lines of the loaded data 
dft.head(100)
```

```python
#initializing constants
from PIL import Image
import numpy as np
import random
import cv2
import matplotlib.pyplot as plt
testnum=105
fn = dft['filename']
Y_train = dft[['xtrans','ytrans']]
area00=(0,0,800,600)
area2=(0,600,800,1200)
img = Image.open(('MoreData/'+ fn[testnum])).convert('L')

#example images that are pasted together in each file
IMA=(np.array(img.crop(area00))) 
im=plt.imshow(IMA,origin="upper",interpolation="nearest",cmap=plt.cm.gray_r)
plt.colorbar(im)      
plt.show()

IMB=(np.array(img.crop(area2))) 
im=plt.imshow(IMB,origin="upper",interpolation="nearest",cmap=plt.cm.gray_r)
plt.colorbar(im)      
plt.show()

Y_train.iloc[testnum]
```

```python
#set up basic indexing
numsamps=200
trainli = np.zeros(numsamps)
testli = np.zeros(numsamps)
Y_train = np.zeros([numsamps,2])
Y_test = np.zeros([numsamps,2])
for lf in range(numsamps):
    #trainli[lf]= random.randint(1,len(dft))
    #testli[lf]= random.randint(1,len(dft))
    trainli[lf]= lf
    testli[lf]= lf
X_train=np.zeros([numsamps, 60,40])
Xn_train=np.zeros([numsamps, 60,40])
X_test=np.zeros([numsamps, 60,40])
Xn_test=np.zeros([numsamps, 60,40])
ig1 = np.zeros([30,40])
ig2 = np.zeros([60,40])
```

```python
#test
#maybe I should define function-- these lines should be identical for test and train
FFT = np.zeros(IMB.shape)[300-20:300+20,400-30:400+30]
for lf in range(len(trainli)):
#train data ORIG
    lft = int(trainli[lf])
    Y_train[lf,:]=dft.iloc[lft].values[8:10]
    img = Image.open(('MoreData/'+ fn[lft])).convert('L')
    IMA = (np.array(img.crop(area00)))        
    IMB = (np.array(img.crop(area2)))
    ig1 = cv2.resize((np.abs(IMA)), dsize=(40,30), interpolation=cv2.INTER_CUBIC)
    X_train[lf,0:30,:] = ig1    
    ig1 = cv2.resize((np.abs(IMB)), dsize=(40,30), interpolation=cv2.INTER_CUBIC)
    X_train[lf,30:60,:] = ig1    
#train data FFT
    FFT = np.array(np.fft.fftshift(np.fft.fft2(IMA+IMB)))[300-20:300+20,400-30:400+30]#why column major? 
    #ig2 = cv2.resize(np.log(np.abs(FFT)), dsize=(40,60), interpolation=cv2.INTER_CUBIC)
    Xn_train[lf,:,:] = np.log(np.transpose(np.abs(np.real(FFT))))
#test data ORIG
    lft = int(testli[lf])
    Y_test[lf,:]=dft.iloc[lft].values[8:10]
    img = Image.open(('MoreData/'+ fn[lft])).convert('L')
    IMA = (np.array(img.crop(area00)))        
    IMB = (np.array(img.crop(area2)))
    ig1 = cv2.resize((np.abs(IMA)), dsize=(40,30), interpolation=cv2.INTER_CUBIC)
    X_test[lf,0:30,:] = ig1
    ig1 = cv2.resize((np.abs(IMB)), dsize=(40,30), interpolation=cv2.INTER_CUBIC)
    X_test[lf,30:60,:] = ig1  
#test data FFT
    FFT = np.array(np.fft.fftshift(np.fft.fft2(IMA+IMB)))[300-20:300+20,400-30:400+30]    
    #ig2 = cv2.resize(np.log(np.abs(FFT)), dsize=(40,60), interpolation=cv2.INTER_CUBIC)
    Xn_test[lf,:,:] = np.log(np.transpose(np.abs(np.real(FFT))))
```

```python
from keras.utils.np_utils import to_categorical 
from keras.callbacks import ReduceLROnPlateau
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.optimizers import RMSprop
#from keras.preprocessing.image import ImageDataGenerator


for l in range(0,6):
    model = Sequential()
    model.add(Dense(2**(l+1), input_shape=(60,40), activation='relu'))
    model.add(Flatten())
    #model.add(Dense(4, activation='relu'))
    #model.add(Dense(2**(l+1), activation='relu'))
    model.add(Dense(2, activation='sigmoid'))
    model.compile(optimizer='rmsprop',loss='categorical_crossentropy', metrics=['accuracy'])

    #optimizer = RMSprop(lr=0.001, rho=0.8, epsilon=1e-10, decay=0.0)
    #model.compile(optimizer = optimizer , loss = "categorical_crossentropy", metrics=["accuracy"])
    #model.compile(optimizer = 'sgd', loss='categorical_crossentropy')
    epochs = 30
    batch_size = 30
    learning_rate_reduction = ReduceLROnPlateau(monitor='val_acc', patience=3,verbose=1,factor=0.5,min_lr=0.00001)

    model.fit(X_train/255, Y_train, epochs=30, verbose=0,shuffle=False)
    finalO = model.fit(X_test/255, Y_test, batch_size = batch_size, epochs = epochs, verbose=0, validation_data = (X_train/255, Y_train),callbacks=[learning_rate_reduction])
    model.fit(Xn_train, Y_train, epochs=100, verbose=0,shuffle=False)
    finalX = model.fit(Xn_test, Y_test, batch_size = batch_size, epochs = epochs, verbose=0, validation_data = (Xn_train, Y_train),callbacks=[learning_rate_reduction])

    plt.plot(finalO.history['acc'])
    plt.plot(finalO.history['val_acc'])
    plt.plot(finalX.history['acc'])
    plt.plot(finalX.history['val_acc'])
    plt.title('accuracy with '+str(2**(l+1)) +' neurons')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['Orig train', 'Orig validation','FFT train', 'FFT validation'], loc='lower left')
    plt.show()
# "Loss"
'''plt.plot(finalO.history['loss'])
plt.plot(finalO.history['val_loss'])
plt.plot(finalX.history['loss'])
plt.plot(finalX.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['O train', 'O validation', 'X train', 'X validation'], loc='lower right')
plt.show()
'''
```
