```python
import pandas as pd 
#dft = pd.read_csv("MoreData/dft.csv",sep='\t') 
dft = pd.read_csv("Commit/dfts.csv",sep='\t') 
# Preview the first 5 lines of the loaded data 
dft.head(100)
```

```python
#initializing constants plot the data
from PIL import Image
import numpy as np
import random
import cv2
import matplotlib.pyplot as plt
testnum=105
fn = dft['filename']
Y_train = dft[['xtrans','ytrans']]
area00=(0,0,800,600)
area2=(0,600,800,1200)
img = Image.open(('MoreData/'+ fn[testnum])).convert('L')

#example images that are pasted together in each file and offset from Y_train
IMA=(np.array(img.crop(area00))) 
im=plt.imshow(IMA,origin="upper",interpolation="nearest",cmap=plt.cm.gray_r)
plt.colorbar(im)      
plt.show()

IMB=(np.array(img.crop(area2))) 
im=plt.imshow(IMB,origin="upper",interpolation="nearest",cmap=plt.cm.gray_r)
plt.colorbar(im)      
plt.show()

Y_train.iloc[testnum]
```

```python
#set up basic indexing
numsamps=50
trainli = np.zeros(numsamps)
testli = np.zeros(numsamps)
Y_train = np.zeros([numsamps,2])
Y_test = np.zeros([numsamps,2])
for lf in range(numsamps):
    #trainli[lf]= random.randint(1,len(dft))
    #testli[lf]= random.randint(1,len(dft))
    trainli[lf]= lf #for now just take first 50
    testli[lf]= lf#for now just take first 50
X_train=np.zeros([numsamps, 60,40])
Xn_train=np.zeros([numsamps, 60,40])
X_test=np.zeros([numsamps, 60,40])
Xn_test=np.zeros([numsamps, 60,40])
ig1 = np.zeros([30,40])
ig2 = np.zeros([60,40])
```

```python
#test
#maybe I should define function-- these lines should be identical for test and train
FFT = np.zeros(IMB.shape)[300-20:300+20,400-30:400+30]
for lf in range(len(trainli)):
#train data ORIG
    lft = int(trainli[lf])
    Y_train[lf,:]=dft.iloc[lft].values[8:10]
    img = Image.open(('MoreData/'+ fn[lft])).convert('L')
    IMA = (np.array(img.crop(area00)))        
    IMB = (np.array(img.crop(area2)))
    ig1 = cv2.resize((np.abs(IMA)), dsize=(40,30), interpolation=cv2.INTER_CUBIC)
    X_train[lf,0:30,:] = ig1    
    ig1 = cv2.resize((np.abs(IMB)), dsize=(40,30), interpolation=cv2.INTER_CUBIC)
    X_train[lf,30:60,:] = ig1    
#train data FFT
    FFT = np.array(np.fft.fftshift(np.fft.fft2(IMA+IMB)))[300-20:300+20,400-30:400+30]#why column major? 
    #ig2 = cv2.resize(np.log(np.abs(FFT)), dsize=(40,60), interpolation=cv2.INTER_CUBIC)
    Xn_train[lf,:,:] = np.log(np.transpose(np.abs(np.abs(FFT))))
#test data ORIG
    lft = int(testli[lf])
    Y_test[lf,:]=dft.iloc[lft].values[8:10]
    img = Image.open(('MoreData/'+ fn[lft])).convert('L')
    IMA = (np.array(img.crop(area00)))        
    IMB = (np.array(img.crop(area2)))
    ig1 = cv2.resize((np.abs(IMA)), dsize=(40,30), interpolation=cv2.INTER_CUBIC)
    X_test[lf,0:30,:] = ig1
    ig1 = cv2.resize((np.abs(IMB)), dsize=(40,30), interpolation=cv2.INTER_CUBIC)
    X_test[lf,30:60,:] = ig1  
#test data FFT
    FFT = np.array(np.fft.fftshift(np.fft.fft2(IMA+IMB)))[300-20:300+20,400-30:400+30]    
    #ig2 = cv2.resize(np.log(np.abs(FFT)), dsize=(40,60), interpolation=cv2.INTER_CUBIC)
    Xn_test[lf,:,:] = np.log(np.transpose(np.abs(np.abs(FFT))))
```
